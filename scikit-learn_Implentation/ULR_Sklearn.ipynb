# Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Importing the dataset

```
# Training data
dataset = pd.read_csv('train.csv')
X_train = dataset.iloc[:, 0].values
y_train = dataset.iloc[:, 1].values
```

```
# Test data
dataset = pd.read_csv('test.csv')
X_test = dataset.iloc[:, 0].values
y_test = dataset.iloc[:, 1].values
```

# Visualizing the data

```
# Train data
plt.scatter(X_train, y_train, color='violet', edgecolor='black')
plt.show()
```

```
# Test data
plt.scatter(X_test, y_test, color='yellow', edgecolor='black')
plt.show()
```

# Preprocessing

```
# making sure that array are of proper shape i.e 2D array
X_train = X_train.reshape((len(X_train), 1))
X_test = X_test.reshape((len(X_test), 1))
y_train = y_train.reshape((len(y_train), 1))
y_test = y_test.reshape((len(y_test), 1))
```

# Training the model on the Training set

```
model = LinearRegression()
model.fit(X_train, y_train)
```

# Plotting the Predictions

```
# Training set
y_pred_train = model.predict(X_train)

plt.scatter(X_train, y_train, color='yellow', edgecolor='black')
plt.plot(X_train, y_pred_train, color='red')
plt.show()
```

```
# Test set
y_pred_test = model.predict(X_test)

plt.scatter(X_test, y_test, color='yellow', edgecolor='black')
plt.plot(X_test, y_pred_test, color='red')
plt.show()
```

# Prediction on a single example

```
x = 300.5

y_pred = model.predict([[x]])
print("The predicted value for x={} is {}".format(x, np.squeeze(y_pred)))
```
